&#9989; **Question 3.2 (4 point):** Perform a PCA of the raisin data in Part 2. Take the first 2 PCAs and run logistic regression using these 2 PCAs. Print out the accuracy.

````python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
### Put your code here

data = pd.read_csv('Raisin_Dataset.csv')
# Drop the 'Area' column
data = data.drop(columns=['Area'])

# Convert class labels to numeric (0 and 1)
data['Class'] = data['Class'].astype('category').cat.codes

# Separate features and labels
X = data.drop(columns='Class')
y = data['Class']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA and keep first 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.25, random_state=42)

# Fit logistic regression on PCA components
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy using first 2 PCA components: {accuracy:.4f}")
````

````python
import random 
import numpy as np
import math
import matplotlib.pyplot as plt
import time  
from IPython.display import display, clear_output
````

````python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', \
                            frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
````

**Human class**

````python
class human():
    
    ## initiate and parameters
    def __init__(self, roam_range, initial_x, initial_y, \
                 lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path
        self.x = initial_x
        self.y = initial_y    

    ## draw images  
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming  
    def roam(self):  
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy        
````

````python
class zombie():
    
    ## initiate and parameters
    def __init__(self, roam_range, hunt_radius, kill_rate, \
                 initial_x, initial_y, lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.radius = hunt_radius
        self.kr = kill_rate
        self.x = initial_x
        self.y = initial_y
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path

    ## draw images    
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming   
    def roam(self):
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy

    ## zombie hunts human
    def hunt(self, human):
        ## part I
        self.h_x = human.x
        self.h_y = human.y

        ## Part II
        dst = math.sqrt((self.x-self.h_x)**2 + (self.y-self.h_y)**2)
        
        catch = False 
        ## Part III
        if dst < self.radius:
            if random.random() < self.kr:
                catch = True
        return catch
        
    ## Part IV    
    def infect(self):
        new_zombie = zombie(self.range, self.radius, self.kr, \
                            self.x, self.y, self.lw, self.up, \
                            self.image_path) 
        return new_zombie
      
````
